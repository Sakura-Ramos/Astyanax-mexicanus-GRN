```python
net =  pd.concat([homo,mus,dmel])
net.shape
```

```python
net.head()
```

```python
net = net.drop(columns=['ID_source', 'ID_target', 'Grupo_interaccion'], errors='ignore')
net = net.rename(columns={'Source_ortolog': 'Source', 'Target_ortolog': 'Target'})
```

```python
net.shape
```

```python
net.head()
```

```python
def resolver_interaccion(grupo):
    # Filtrar interacciones conocidas
    conocidos = grupo[grupo['Interaction'].isin(['+', '-'])]
    if not conocidos.empty:
        return conocidos.iloc[[0]]  # Conserva una interacción conocida
    else:
        return grupo.iloc[[0]]  # Si no hay conocidas, conserva una con '?'

# Aplicar la función agrupando por Source y Target
global_net_filtrado = net.groupby(['Source', 'Target'], group_keys=False).apply(resolver_interaccion).reset_index(drop=True)
```

```python
global_net_filtrado.to_csv('CF_global.csv')
```
