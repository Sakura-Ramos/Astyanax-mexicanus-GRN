#### Redes modelo en base a datos de ortología ###
1. Datos 
2. Filtrar las interacciones para asegurar que source y target sean ortologos de astyanax 
3. Pasar de nombres de genes del organismo modelo a ID's de Astyanax 

## Pachon - Dmel

```python
# Filtramos ortólogos de 'Source'
source = dmel[dmel['Source'].isin(orto_dmel['b2'])]

# Filtramos ortólogos de 'Target' dentro del subconjunto filtrado previamente
net_dmel = source[source['Target'].isin(orto_dmel['b2'])]

# Eliminamos duplicados, manteniendo solo la primera ocurrencia
net_dmel = net_dmel.drop_duplicates()

# Crear diccionario de mapeo de 'b2' a 'a'
mapp = dict(zip(orto_dmel['b2'], orto_dmel['a']))

# Reemplazar valores en 'Source' y 'Target' usando el mapeo de 'b2' → 'a'

net_dmel.replace({'Source': mapp, 'Target': mapp}, inplace=True)

# Posibles duplicados tras el mapeo
net_dmel = net_dmel.drop_duplicates()
```
