```python
import pandas as pd
import numpy as np
import networkx as nx
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Datos
net = pd.read_csv("/content/drive/My Drive/Sakura/Subnet_SF.csv")
net=net[['Source','Target']]
net.head()
```

```python
len(net)
```

```python
# Transformamos la lista de interacciones en un grafo
Net = nx.from_pandas_edgelist(net, 'Source', 'Target', create_using=nx.DiGraph())
```

```python
# Observamos la información basica del grafo
print('Number of nodes', len(Net.nodes))
print('Number of edges', len(Net.edges))
```

```python
#list(Net.nodes(data=True))
```

```python
path = nx.shortest_path(Net, target='SIX6B')
path
```

```python
# 4. Convertir 'path' en un DataFrame
df_path = pd.DataFrame.from_dict(path, orient='index')

# Rename columns to 'Camino_0', 'Camino_1', etc.
df_path.columns = ['Camino_' + str(i) for i in range(df_path.shape[1])]
```

```python
df_path.to_csv('df_path.csv', index=False)
from google.colab import files
files.download('df_path.csv')
```

```python
#Datos
net = pd.read_csv("/content/drive/My Drive/Sakura/Subnet_Pachon.csv")
net=net[['Source','Target']]
net.head()
```

```python
len(net)
```

```python
# Transformamos la lista de interacciones en un grafo
G = nx.from_pandas_edgelist(net, 'Source', 'Target', create_using=nx.DiGraph())
```

```python
# Observamos la información basica del grafo
print('Number of nodes', len(G.nodes))
print('Number of edges', len(G.edges))
```

```python
Nodes = list(G.nodes(data=True))
```

```python
df_nodes = pd.DataFrame.from_dict(Nodes)
df_nodes
```

```python
df_nodes.to_csv('df_nodes.csv', index=False)
from google.colab import files
files.download('df_nodes.csv')
```
