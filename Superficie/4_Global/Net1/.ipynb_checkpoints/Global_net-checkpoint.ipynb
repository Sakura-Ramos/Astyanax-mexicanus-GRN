```python
mus = pd.read_csv('Ast_Netw_from_Mus.csv')
mus.head()
```

```python
dmel = pd.read_csv('Ast_Netw_from_Dmel.csv')
dmel.head()
```

```python
dmel.shape
```

```python
net =  pd.concat([homo,mus,dmel])
```

```python
net.shape
```

```python
# Eliminar filas duplicadas
global_net = net.drop_duplicates()
```

```python
global_net.shape
```

```python
global_net
```

```python
def resolver_interaccion(grupo):
    # Filtrar interacciones conocidas
    conocidos = grupo[grupo['Interaction'].isin(['+', '-'])]
    if not conocidos.empty:
        return conocidos.iloc[[0]]  # Conserva una interacción conocida
    else:
        return grupo.iloc[[0]]  # Si no hay conocidas, conserva una con '?'

# Aplicar la función agrupando por Source y Target
global_net_filtrado = global_net.groupby(['Source', 'Target'], group_keys=False).apply(resolver_interaccion).reset_index(drop=True)
```

```python
global_net_filtrado.shape
```

```python
global_net_filtrado.to_csv('Global_net.csv')
```
