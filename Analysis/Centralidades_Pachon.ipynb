```python
#Datos
net_Pach = pd.read_csv("/content/drive/My Drive/Sakura/Subnet_Pachon.csv")
net_Pach=net_Pach[['Source','Target']]
net_Pach.head()
```

```python
len(net_Pach)
```

```python
G = nx.from_pandas_edgelist(net_Pach, 'Source', 'Target', create_using=nx.DiGraph)
```

```python
print(nx.is_directed(G))
```

```python
# Observamos la informaci√≥n basica del grafo
print('Number of nodes', len(G.nodes))
print('Number of edges', len(G.edges))
```

```python
import operator
DegreeCen=nx.degree_centrality(G)
sortedDCen = sorted(DegreeCen.items(), key=operator.itemgetter(1), reverse=True)
for d in sortedDCen[:20]:
    print(d)
```

```python
betweenness_dict = nx.betweenness_centrality(G) # Run betweenness centrality
nx.set_node_attributes(G, betweenness_dict, 'betweenness')
```

```python
sorted_betweenness = sorted(betweenness_dict.items(), key=operator.itemgetter(1), reverse=True)

print("Top 20 nodes by betweenness centrality:")
for b in sorted_betweenness[:20]:
    print(b)
```

```python
closeness_dict = nx.closeness_centrality(G)
sorted_closeness = sorted(closeness_dict.items(),key=operator.itemgetter(1), reverse=True)

f = open("closeness_centrality.txt","w")
f.write( str(closeness_dict) )
f.close()

for b in sorted_closeness[:20]:
    print(b)
```
